```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/library_data.csv")
```

```python
df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df = df.loc[:, ~df.columns.duplicated()]
```

```python
[col for col in df.columns if "type" in col or "book" in col]
```

```python
import matplotlib.pyplot as plt

type_for_book = df["preferred_type_for_reading"].value_counts()

type_for_book.head(5).plot(kind="bar")
plt.title("Top 5 Preferred Types of Books")
plt.xlabel("Type of Books")
plt.ylabel("Number of Students")
plt.show()
```

## Final Insights

1. The analysis shows that certain types of books are preferred more by students compared to others.
2. Academic and study-related books have higher preference, indicating that students mainly use reading resources for educational purposes.
3. Non-academic reading preferences are comparatively lower, suggesting limited leisure reading habits among students.
4. The distribution of preferred book types highlights studentsâ€™ focus on curriculum-oriented learning.
5. These insights can help the college library improve book availability based on student preferences.
